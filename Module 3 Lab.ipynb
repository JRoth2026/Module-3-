```python
# File Name: Module 3 Lab
# Author: Jack Roth
# Description: This code uses classes and inheritance in python. The car super class stores the vehicle type and the subclass inherits from the vehicle
# and saves information about the the car's year, make, model, number of doors, and roof type. This code takes input and then shows the vehicle information
# in a readable format.
# Variables:
# vehicle_type - Stores the type of car
# year - Stores the year
# make - Stores the manufacturer
# model - Stores the model
# doors - Stores the number of doors (2 or 4)
# roof - Stores the roof type (solid or sun roof)
# my_car - Car object containing all the information

# Super Class
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

# Sub Class
class Automobile(Vehicle):
    def __init__(self, vehicle_type, year, make, model, doors, roof):
        super().__init__(vehicle_type)

        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof
        
# User Input
year = input("Enter the year: ")
make = input("Enter the make: ")
model = input("Enter the model: ")
doors = input("Enter the number of doors (2 or 4): ")
roof = input("Enter the roof type (solid or sun roof): ")

# Create Object
my_car = Automobile("car", year, make, model, doors, roof)

# Output
print("\nVehicle Information")
print("-------------------")
print("Vehicle type:", my_car.vehicle_type)
print("Year:", my_car.year)
print("Make:", my_car.make)
print("Model:", my_car.model)
print("Number of doors:", my_car.doors)
print("Type of roof:", my_car.roof)
```
